```python
def magnitude(ax , ay):
    s = sqrt(ax**2 + ay**2)
    return s
```

```python
magnitude(3,2)
```

```python
def components(mag,theta):
    angle_rad =radians(theta) 
    ax = mag * cos(angle_rad)
    ay = mag * sin(angle_rad)
    return ax, ay

components(4, 45)
```

```python
def angle(ax, ay):
     t = atan(ay/ax)
     theta = degrees(t)
     return theta
angle(3,3)
```

# Plotting in Python using Matplotlib  

Matplotlib is an excellent 2D and 3D graphics  library for generating scientific figures based on NumPy. Some of the many advantages of this library include:


*	Easy to get started
*	Support for formatted labels and texts 
*	Great control of every element in a figure, including figure size and DPI. 
*	High-quality output in many formats, including PNG, PDF, SVG, EPS, and PGF.
*	 GUI for interactively exploring figures and support for headless generation of figure files (useful for batch jobs).




```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot([1,2,3,4])
plt.show()
```

**Plot is a versatile function, and will take an arbitrary number of arguments. For example, to plot x versus y, you can write:** 

```python
plt.plot([1,2,3,4],[2,13,44,55])
plt.show()
```

## Formatting the style of your plot

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot. You concatenate a color string with a line style string. The default format string is 'b-', which is a solid blue line. For example, to plot the above with red circles, you would issue.


```python
plt.plot([1,2,3,4,5],[12,34,56,66,72],'bo')     # bo is blue circle pattern 
plt.show() 
```

```python
plt.plot([1,2,3,4,5],[12,34,56,66,72],'r-')     # r- is red solid line
plt.show() 
```

## Numpy in matplotlib
If matplotlib were limited to working with lists, it would be fairly useless for numeric processing. Generally, you will use numpy arrays. In fact, all sequences are converted to numpy arrays internally. The example below illustrates plotting several lines with different format styles in one function call using arrays.

```python
%matplotlib inline 
import numpy as np
t = np.arange(0.0,5,0.1)
plt.plot(t,t, 'r--' )
plt.show()
```

```python
stp = np.arange(0.,10,1)
plt.plot(stp,stp**2, 'g--', stp,stp**3, 'bs' , stp,stp**4, 'r^')
plt.show()
```

```python
x= np.arange(0,2*np.pi,0.1 )
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x= np.arange(0,2*np.pi,0.1 )
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x= np.arange(0,2*np.pi,0.1 )
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x,sin, 'r', x,cos , 'b')
plt.show()
```

```python
x= np.arange(0,4*np.pi,0.1 )
cos = np.cos(x)
plt.plot(x,cos, 'r', x,-cos , 'b')
plt.show()
```

```python
plt.plot(x,cos,linewidth= 2,linestyle='--', color='g')
plt.show()
```
