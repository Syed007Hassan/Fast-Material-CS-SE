```python
import numpy as np
```

```python
def projectile(vo,t):
    angle=np.radians(t)
    H=((vo**2))*(np.sin(angle))**2/(2*(9.8))
    R=((vo**2))*(np.sin(2*angle))/9.8
    return H,R
```

```python
projectile(10,48)
```

### Q3.


```python
def SHM(x, w, q, t):
    w =np.radians(w)
    ph =np.radians(q)
    x=x*np.cos((w*t)+ph)
    v=(-x)*(w)*(np.sin((w*t)+ph))
    a=(x)*(w)*(2*np.cos((w*t)+ph))
    return x, v, a
SHM(25,  45, 30, 4)
    
```

```python
import matplotlib.pyplot as plt
x=np.arange(0,np.pi/3,0.0001)
y=10*(np.cos((60*x)+45))
plt.plot(x,y)
plt.title('Displacement with respect to time')
plt.show
```
