## Linear Independence

A set of n vectors of length n is linearly independent if the matrix with these vectors as columns has a non-zero determinant

$$
\left[\begin{matrix}x\\y\\z\end{matrix}\right]=s
\left[\begin{matrix}1\\-1\\1\end{matrix}\right]+t
\left[\begin{matrix}1\\1\\1\end{matrix}\right]=
\left[\begin{matrix}s+t\\-s+t\\s+t\end{matrix}
\right]
$$

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(projection='3d')

s = np.linspace(-1, 1, 10)
t = np.linspace(-1, 1, 10)
S, T = np.meshgrid(s, t)

X = S+T
Y = -S+T
Z = S+T
ax.plot_wireframe(X, Y, Z, linewidth = 1.5, color = 'yellow', alpha = .6)

vec = np.array([[[0, 0, 0, 2, -1, 3]],#red
               [[0, 0, 0, 1, 1/2, 1]],#blue
               [[0, 0, 0, -1, 1, 3]]])#green
colors = ['r','b','g']
for i in range(vec.shape[0]):
    X, Y, Z, U, V, W = zip(*vec[i,:,:])
    #for setting arrows
    ax.quiver(X, Y, Z, U, V, W, length=1, normalize=False, color = colors[i],
              arrow_length_ratio = .1, pivot = 'tail',
              linestyles = 'solid',linewidths = 3, alpha = .6)

ax.set_title('Linear Independence Visualization')

ax.set_xlabel('x-axis', size = 12)
ax.set_ylabel('y-axis', size = 12)
ax.set_zlabel('z-axis', size = 12)

ax.view_init(elev=50., azim=0)
plt.show()
```

We can see that the greenvector is not in the plane spanned red and blue vector, thus they are linearly independent.

### Finding the dimension and bases of Row space and Column space

The dimension of the row space is called the rank of the matrix. This is the same as the maximum number of linearly independent rows that can be chosen from the matrix, or equivalently the number of pivots.

The dimension of the column space is also called the rank of the matrix. Here the rank is also equal to the number of pivots in the reduced row echelon form, and is the maximum number of linearly independent columns that can be chosen from the matrix.

```python
A = sy.Matrix([[1, 8, 13, 12, 1],
               [14, 11, 2, 7, 2], 
               [4, 5, 26, 9, 1], 
               [15, 10, 3, 6, 3]]);A
```

```python
B = A.rref();B
```

The basis of row space of $A$ is its first 4 rows: $(1,8,13,12,1), (14,11,2,7,2), (4,5,26,9,1), ( 15,10,3,6,3) $ So dimension of row space is $4$. 
The basis of column space of $A$ is its first 4 cols: $(1,14,4,15), (8,11,5,10), (13,2,26,3), (12,7,9,6) $ So dimension of col space is also $4$.

### Eigen value and Eigen Vector

Let 

$$
A= 
\left[
\begin{matrix}
1 & 8 & 9 \\
5 & 9 & 3 \\
3 & 6 & -1 \\
\end{matrix}
\right]
$$


Use characteristic equation $|A - \lambda I|=0$

$$
\left|
\left[
\begin{matrix}
1 & 8 & 9 \\
5 & 9 & 3 \\
3 & 6 & -1 \\
\end{matrix}
\right]
-
\left[
\begin{matrix}
\lambda & 0 & 0\\ 
0 & \lambda & 0\\
0 & 0 & \lambda
\end{matrix}
\right]\right|=0
$$

```python
lamda = sy.symbols('lamda')
A = sy.Matrix([[1,8,9],[5,9,3], [3,6,-1]])
I = sy.eye(3)
```

```python
A - lamda*I
```

```python
p = A.charpoly(lamda);p
```
